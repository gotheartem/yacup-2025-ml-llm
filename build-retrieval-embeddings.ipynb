```python
from llama_cpp import Llama

llm = Llama(
    model_path="./models/Qwen3-Embedding-0.6B-GGUF/Qwen3-Embedding-0.6B-Q8_0.gguf",
    embedding=True,
    n_gpu_layers=-1,
    verbose=False,
)
```

```python
import json
with open('data/articles.json', 'r') as file:
    texts = json.load(file)
```

```python
import sys, os
import numpy as np
from tqdm.notebook import tqdm as tqdmn

with open(os.devnull, 'w') as file:
    o = sys.stderr
    sys.stderr = file
    embeds = []
    for s in tqdmn(texts):
        embeds.append(np.array(llm.create_embedding(input=s)["data"][0]["embedding"], dtype=np.float16))
    sys.stderr = o

e = np.stack(embeds)
print(e.shape)

with open('data/embeds.npy', 'wb') as file:
    e.dump(file)
```
